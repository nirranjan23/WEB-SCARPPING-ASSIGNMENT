```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.flipkart.com/search?q=asus+laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_9_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_9_na_na_na&as-pos=2&as-type=RECENT&suggestionId=asus+laptop%7CLaptops&requestId=6102646a-f1b4-492e-9be4-2942dd83647e&as-searchtext=laptops%20a"
```

```python
permission=requests.get(url)
```

```python
permission
```

```python
LAPTOP=BeautifulSoup(permission.text)
```

```python
LAPTOP
```

```python
LAPTOP_NAME=LAPTOP.select("._4rR01T")
LAPTOP_WINDOWS=LAPTOP.select(".rgWa7D:nth-child(3)")
LAPTOP_SSD=LAPTOP.select(".rgWa7D:nth-child(4)")
LAPTOP_INCH=LAPTOP.select(".rgWa7D:nth-child(5)")
```

```python
import pandas as py
import numpy as np
```

```python
details=py.DataFrame(LAPTOP_NAME,columns=['LAPTOP'])
details['window']=LAPTOP_WINDOWS
details['storage']=LAPTOP_SSD
details['size']=LAPTOP_INCH
details
```

```python
details.to_csv('C:/web scrapping/laptop.csv')
```
